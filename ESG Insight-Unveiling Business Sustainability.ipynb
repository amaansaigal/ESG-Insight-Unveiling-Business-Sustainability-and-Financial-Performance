## Visualizations for businesss dataset



#### Total Energy Consumption by Sector in the UK

In this section, we will explore and analyze the total energy consumption across various sectors in the United Kingdom. Understanding the distribution of energy consumption by sector is crucial for assessing resource allocation and sustainability efforts.

Our analysis aims to shed light on the relative energy demands of different sectors within the UK, providing valuable insights into the nation's energy landscape.


```python
df_energy_consumption = df_esg_business.groupby(['Sector'])[['Coal_Consumption_By_Sector(toe)', 'Electricity_Consumption_By_Structure(toe)', 'Gas_Consumption_By_Structure(toe)',\
                                     'NG_Consumption_By_Sector(toe)', 'Petrol_Consumption_By_Sector(toe)' ]].sum()
```

```python
#convert each value to percentage as it can be easily scaled and comparision will be good.

for col in df_energy_consumption:
    df_energy_consumption[col] = (df_energy_consumption[col]/df_energy_consumption[col].sum())*100
    
df_energy_consumption
```

```python
ax = df_energy_consumption.plot( kind='barh', stacked= True, color=['red', 'blue', 'green','violet', 'orange'])
ax.legend(bbox_to_anchor = (1.05,1))
plt.figure(figsize=(40,20))
```

## Key Insights from the Energy Consumption Analysis

Based on the figure presented above, we can draw several notable conclusions regarding energy consumption patterns across different sectors in the UK:

1. **Highest Energy Consumption in the "Production" Sector:** The manufacturing or "Production" sector stands out as the highest consumer of energy among all sectors. This indicates its significant energy demands for various industrial processes.

2. **Negligible Coal Usage Across Most Industries:** Coal consumption appears to be negligible in the majority of industries, except for the "Production" sector. This suggests a substantial reliance on alternative energy sources in most sectors.

3. **Predominance of Electricity Consumption in Other Sectors:** In sectors apart from "Production," electricity consumption takes precedence as the primary energy source. This signifies a reliance on electricity for powering various operations and processes.

4. **Professional and Science Sector Leads in Gas Consumption:** The "Professional and Science" sector emerges as the leading consumer of gas, indicating a substantial requirement for gas-related activities within this sector.

5. **Natural Gas Consumption Dominance in Production Sector:** Similar to overall energy consumption, the "Production" sector also exhibits the highest natural gas consumption. This highlights its dependence on natural gas for industrial operations.

6. **Petrol Consumption Peaks in Construction and Retail Sectors:** Among all sectors, the "Construction" and "Retail" sectors showcase the highest petrol consumption. This aligns with their extensive transportation needs for goods and materials, contributing to higher petrol usage.

These insights offer a comprehensive overview of energy consumption trends, providing valuable information for decision-makers and policymakers in the energy and industrial sectors.


## Energy distribution by Structure type

```python
#convert each value to percentage as it can be easily scaled and comparision will be good.

for col in df_esg_business:
    if df_esg_business[col].dtype == float:
        df_esg_business[col] = (df_esg_business[col]/df_esg_business[col].sum())*100
    
df_esg_business.head()
```

```python
# Lets create a column which has total energy consumption by each row.

float_cols=[]
for col in df_esg_business:
    if df_esg_business[col].dtype == float:
        float_cols.append(col)
        
        
df_esg_business['total_energy_consumed'] = df_esg_business[float_cols].sum(axis=1)
```

```python
df_esg_business.groupby('Structure_Type')['total_energy_consumed'].sum().plot(kind = 'bar')
```

## Inference: Factories as Major Contributors to Pollution

A significant inference drawn from the analysis is that the highest energy consumption is primarily attributed to factories or the "Production" sector. This observation leads to a compelling conclusion: factories play a substantial role in contributing to environmental pollution.

The correlation between high energy consumption and pollution is well-established, as factories often rely on energy-intensive processes, machinery, and production methods. Consequently, this sector's prominence in energy consumption strongly suggests that it is a major source of pollution.

Addressing and mitigating pollution from factories should thus be a critical focus for environmental initiatives and regulatory measures. By recognizing the impact of industrial energy consumption, we can take proactive steps toward reducing pollution and promoting sustainability.


```python
# Now lets see the top 20 companies which consume the highest energy.

fig, ax = plt.subplots(figsize = (10,9))

df_temp = df_esg_business.sort_values('total_energy_consumed', ascending=False)[['id','total_energy_consumed']].head(20)
sns.barplot(x = 'id', y='total_energy_consumed', data= df_temp, order = df_temp.sort_values('total_energy_consumed', ascending= False).id)
```

## Key Observations:

Two noteworthy observations can be made from the analysis:

1. **Highest Fuel Consumption by Company ID 420:** The data reveals that Company ID 420 stands out as the highest consumer of fuel among all companies. This indicates that Company ID 420 has a significant fuel requirement compared to others in the dataset.

2. **Dominance of the 'Production' Sector:** It's notable that the vast majority of companies in the dataset belong to the 'Production' sector. This suggests a prevalence of companies engaged in manufacturing or industrial activities, which might explain the high overall fuel consumption.


```python
# Lets also see the top 20 places which consume the maximum energy in UK

df_temp = df_esg_business.groupby('Local Authority')['total_energy_consumed'].sum().sort_values(ascending = False).head(20)
df_temp.sort_values(ascending = False).plot(kind = 'barh', figsize = (10,9))
```
